```python
from matplotlib.patches import Rectangle
```

```python
def flip_h(img, xmin=None, ymin=None, xmax=None, ymax=None,):
    # horiental flip with boundbox
    img_flip_h = img[:, ::-1]
    
    if xmin==None:
        return img_flip_h
    
    else:
        A = img.shape[0] / 2   # find the flip axis
        xmin_new = xmin + 2*( A - xmin)
        ymin_new = ymin
        xmax_new = xmax + 2*( A - xmax)
        ymax_new = ymax
        return img_flip_h, xmin_new, ymin_new, xmax_new, ymax_new


def flip_v(img, xmin=None, ymin=None, xmax=None, ymax=None,):
    # vertical flip with boundbox
    img_flip_v = img[::-1, :]
    if xmin==None:
        return img_flip_v
    else:
        A = img.shape[0] / 2   # find the flip axis
        xmin_new = xmin 
        ymin_new = ymin + 2*( A - ymin)
        xmax_new = xmax 
        ymax_new = ymax + 2*( A - ymax)
        return img_flip_v, xmin_new, ymin_new, xmax_new, ymax_new
```
